```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math 

from scipy.integrate import odeint
```

```python
l_d=0.085
k=1.001
```

```python

n0 = 100
def model(t,n):
    dndt = (k / l_d) * n
    return dndt



t = np.linspace(0,10)


n = odeint(model,n0,t)
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.dpi']= 250
plt.rc('text', usetex = True)
plt.rc('font', family = 'serif')

fig = plt.figure()
plt.plot(t,n,label='Neutron Population')



plt.legend(loc='best')

plt.xlabel('Time (s)')
plt.xlim([0, 10])

plt.ylabel('Population')

plt.title("Simple Point Kinetics Model with delayed neutrons, $k$={}".format(k))
plt.draw()
plt.savefig('SIMPLEPRKdelayed.png', dpi=300)
plt.draw()
plt.show()
```
