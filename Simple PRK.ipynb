```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import math 

from scipy.integrate import odeint
```

```python
l=1e-5
k=1.001
Λ=l/k
ρ=(k-1)/k
```

```python
n0 = 100
def model(t,n):
    dndt = (ρ / Λ) * n
    return dndt



t = np.linspace(0,10)

n = odeint(model,n0,t)
```

```python
plt.style.use('ggplot')
plt.rcParams['figure.dpi']= 250
plt.rc('text', usetex = True)
plt.rc('font', family = 'serif')

fig = plt.figure()
plt.plot(t,n,label='Neutron Population')


plt.legend(loc='best')

plt.xlabel('Time (s)')
plt.xlim([0, 10])

plt.ylabel('Population')
#plt.ylim([0, np.amax(n)+20])

plt.title("Simple Point Kinetics Model, $k$={}".format(k))
plt.draw()
plt.savefig('SIMPLEPRK.png', dpi=300)
plt.draw()
plt.show()
```
